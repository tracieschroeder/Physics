```python
from anastruct import SystemElements
ss1 = SystemElements()
ss2 = SystemElements()
ss3 = SystemElements()
ss4 = SystemElements()
ss5 = SystemElements()
```

Start with a simple beam bridge. Each element has two x,y coordinates, just like points on a graph.

Remember nodes are what we call the points of support. There are four different types that we can build with this library, but we really only need one for this bridge. A hinged node can rotate, but is in a fixed location.

```python
ss1.add_element(location=[[0, 0], [20, 0]]) #The first half of the element. I have to break this up because I don't know how to add a point force to the center of an element.
ss1.add_element(location=[[20, 0], [40, 0]]) #The second half of the element

ss1.add_support_hinged(node_id=1) #support nodes at each end. I'm not clear whether or not this should be a hindged node or a fixed node.
ss1.add_support_hinged(node_id=3)

ss1.point_load(node_id=2, Fy=-10) #There should probably be a way to make this force on an element instead of a node, but I don't know what it is

ss1.show_structure() #This prints your structure
```

```python
#this method compute all the reaction forces and axial forces in the model
ss1.solve()

#this method plot the computed reaction forces at the support
ss1.show_reaction_force()
#this method plot the computed axial forces of the truss elements or members
ss1.show_axial_force()

#this method plot the computed displacement of the entire model.
ss1.show_displacement(factor=15)
```

```python
#this method compute all the reaction forces and axial forces in the model
ss2.solve()

#this method plot the computed reaction forces at the support
ss2.show_reaction_force()
#this method plot the computed axial forces of the truss elements or members
ss2.show_axial_force()
#ss2.show_shear_force()


#this method plot the computed displacement of the entire model.
ss2.show_displacement(factor=15)
```

Now add a truss bridge

```python
ss3.add_element(location=[[0, 0], [20, 0]]) #The first half of the bottom element
ss3.add_element(location=[[20, 0], [40, 0]]) #The second half of the bottom element
ss3.add_element(location=[[0, 20], [20, 20]]) #The first half of the top element
ss3.add_element(location=[[20, 20], [40, 20]]) #The second half of the top element
ss3.add_element(location=[[0, 0], [0, 20]]) #The first vertical element
ss3.add_element(location=[[20, 0], [20, 20]]) #The first vertical element
ss3.add_element(location=[[40, 0], [40, 20]]) #The first vertical element
ss3.add_element(location=[[0, 0], [20, 20]]) #The first up x element
ss3.add_element(location=[[0, 20], [20, 0]]) #The first down x element
ss3.add_element(location=[[20, 0], [40, 20]]) #The second up x element
ss3.add_element(location=[[20, 20], [40, 0]]) #The second down x element


ss3.add_support_hinged(node_id=1) #support nodes at each end. I'm not clear whether or not this should be a hindged node or a fixed node.
ss3.add_support_hinged(node_id=3)

ss3.point_load(node_id=2, Fy=-10) #There should probably be a way to make this force on an element instead of a node, but I don't know what it is3
ss3.show_structure() #This prints your structure
```

```python
#this method compute all the reaction forces and axial forces in the model
ss3.solve()

#this method plot the computed reaction forces at the support
ss3.show_reaction_force()
#this method plot the computed axial forces of the truss elements or members
ss3.show_axial_force()

#this method plot the computed displacement of the entire model.
ss3.show_displacement(factor=15)
```

```python
ss4.add_element(location=[[0, 0], [4, 0]])
ss4.add_element(location=[[0, 0], [2,4]])
ss4.add_element(location=[[4, 0], [8, 0]])
ss4.add_element(location=[[4, 0], [2,4]])
ss4.add_element(location=[[4, 0], [6,4]])
ss4.add_element(location=[[2, 4], [6,4]])
ss4.add_element(location=[[6, 4], [8,0]])

ss4.add_support_hinged(node_id=1)
ss4.add_support_hinged(node_id=4)

ss4.point_load(node_id=3, Fy=-400)
ss4.point_load(node_id=5, Fy=-800)

ss4.show_structure() #This prints your structure
```

```python
ss5.add_element(location=[[0, 0], [3, 0]])
ss5.add_element(location=[[0, 4], [3,0]])

ss5.add_support_hinged(node_id=1)
ss5.add_support_hinged(node_id=3)

ss5.point_load(node_id=2, Fy=-50)

ss5.show_structure() #This prints your structure
```

```python
#this method compute all the reaction forces and axial forces in the model
ss5.solve()

#this method plot the computed reaction forces at the support
ss5.show_reaction_force()
#this method plot the computed axial forces of the truss elements or members
ss5.show_axial_force()

#this method plot the computed displacement of the entire model.
ss5.show_displacement(factor=15)
```

```python
ss5.solve(force_linear=False, verbosity=0, max_iter=200, geometrical_non_linear=False)
```
